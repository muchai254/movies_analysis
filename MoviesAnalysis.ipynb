### Top genres by ratings

```python
movies
```

```python
# Step 1: Filter the dataframe for ratings 7.6 and above
highest_filtered_df = movies[movies['IMDB Rating'] >= 7.6]

# Step 2: Split the genres column
highest_genres_list = highest_filtered_df['Genre'].str.split(',')

# Step 3: Flatten the list of genres
highest_flat_genres_list = [genre.strip() for sublist in highest_genres_list for genre in sublist]
highest_flat_genres_list
```

```python
# Step 4: Count the occurrences of each genre
highest_genre_counts = pd.Series(highest_flat_genres_list).value_counts()

# Step 5: Sort the genres based on the number of occurrences
highest_sorted_genre_counts = highest_genre_counts.sort_values(ascending=False)

# Print or further analyze sorted_genre_counts
print(highest_sorted_genre_counts)
```

```python
# Step 1: Filter the dataframe for ratings 7.6 and above
lowest_filtered_df = movies[movies['IMDB Rating'] <= 6.4]

# Step 2: Split the genres column
lowest_genres_list = lowest_filtered_df['Genre'].str.split(',')

# Step 3: Flatten the list of genres
lowest_flat_genres_list = [genre.strip() for sublist in lowest_genres_list for genre in sublist]
lowest_flat_genres_list
```

```python
# Step 4: Count the occurrences of each genre
lowest_genre_counts = pd.Series(lowest_flat_genres_list).value_counts()

# Step 5: Sort the genres based on the number of occurrences
lowest_sorted_genre_counts = lowest_genre_counts.sort_values(ascending=False)

# Print or further analyze sorted_genre_counts
print(lowest_sorted_genre_counts)
```

### Movies released over the years

```python
# Convert release year to integers
movies['Release Year'] = movies['Release Year'].astype(int)

# Grouping the data by release year and counting the number of movies released each year
movies_per_year = movies.groupby('Release Year').size()

# Plotting the bar graph
plt.figure(figsize=(10, 6))
movies_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Movies Released by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies Released')
tick_frequency = 5  # Label every 5 years, adjust as needed
plt.xticks(range(0, len(movies_per_year), tick_frequency), movies_per_year.index[::tick_frequency], rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Trend of gross income over the years

```python
total_gross_per_year = movies.groupby('Release Year')['Gross'].sum()

# Plotting the bar graph
plt.figure(figsize=(12, 6))
total_gross_per_year.plot(kind='bar', color='skyblue')
plt.title('Total Gross Income by Production Year')
tick_frequency = 5  # Label every 5 years
plt.xticks(range(0, len(total_gross_per_year), tick_frequency), total_gross_per_year.index[::tick_frequency], rotation=45)
plt.ylabel('Gross')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Comparison of gross and production againist years

```python
# Creating subplots to plot both bar graphs on the same figure
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotting the total gross income
color = 'tab:blue'
ax1.set_xlabel('Production Year')
ax1.set_ylabel('Gross', color=color)
ax1.bar(total_gross_per_year.index, total_gross_per_year, color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Creating a secondary y-axis for number of movies
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Number of Movies', color=color)
ax2.bar(movies_per_year.index, movies_per_year, color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Adding titles and labels
plt.title('Comparison of Total Gross Income and Number of Movies by Production Year')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
