```python
movies = pd.DataFrame(data)
```

## longest and shortest movie duration, highest and lowest IMDB rating

```python
movies.describe()
```

```python
movies['Release Year'] = pd.to_numeric(movies['Release Year'], errors='coerce')
```

```python
movies = movies.dropna(subset=['Release Year'])
```

```python
movies.loc[:, "Release Year"].describe()
```

```python
movies.describe()
```

```python
movies.head()
```

```python
movies  = movies.dropna()
movies
```

```python
movies.describe()
```

```python
lowest_duration = movies.loc[:, "Duration"] == 64
movies[lowest_duration]
```

```python
highest_duration = movies.loc[:, "Duration"] == 238
movies[highest_duration]
```

```python
highest_ratings = movies.loc[:, "IMDB Rating"] >= 8
movies[highest_ratings]
```

```python
highest_ratings = movies[movies["IMDB Rating"] >= 8]
highest_ratings_sorted = highest_ratings.sort_values(by="IMDB Rating", ascending=False)
highest_ratings_sorted
```

```python
lowest_ratings = movies[movies["IMDB Rating"] <= 3]
lowest_ratings_sorted = lowest_ratings.sort_values(by="IMDB Rating", ascending=True)
lowest_ratings_sorted
```

```python
# Split the strings in the "genres" column
movies['Genre'] = movies['Genre'].str.split(',')

# Filter the DataFrame for a specific genre
specific_genre = 'Action'  # Change this to your desired genre
action_df = movies[movies['Genre'].apply(lambda x: specific_genre in x)]

# Print out the filtered DataFrame
print(action_df)
```

```python
# Split the strings in the "genres" column
movies['Genre'] = movies['Genre'].str.split(',')

# Filter the DataFrame for a specific genre
specific_genre = 'Crime' 
filtered_df = movies[movies['Genre'].apply(lambda x: specific_genre in x)]

# Print out the filtered DataFrame
print(filtered_df)
```
